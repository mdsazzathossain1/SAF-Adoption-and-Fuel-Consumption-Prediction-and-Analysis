```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Create a DataFrame from your data
data = {
    'Month': [
        '2019-01-01', '2019-02-01', '2019-03-01', '2019-04-01', '2019-05-01', '2019-06-01',
        '2019-07-01', '2019-08-01', '2019-09-01', '2019-10-01', '2019-11-01', '2019-12-01',
        '2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01', '2020-05-01', '2020-06-01',
        '2020-07-01', '2020-08-01', '2020-09-01', '2020-10-01', '2020-11-01', '2020-12-01',
        '2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01', '2021-06-01',
        '2021-07-01', '2021-08-01', '2021-09-01', '2021-10-01', '2021-11-01', '2021-12-01',
        '2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01', '2022-06-01',
        '2022-07-01', '2022-08-01', '2022-09-01', '2022-10-01', '2022-11-01', '2022-12-01',
        '2023-01-01', '2023-02-01', '2023-03-01'
    ],
    'Emissions': [
        5.80E+08, 5.26E+08, 5.39E+08, 5.90E+08, 6.12E+08, 6.24E+08, 6.66E+08, 6.65E+08,
        6.19E+08, 6.23E+08, 5.78E+08, 6.27E+08, 6.06E+08, 4.95E+08, 3.85E+08, 1.34E+08,
        1.63E+08, 1.91E+08, 2.46E+08, 2.80E+08, 2.70E+08, 3.06E+08, 3.02E+08, 3.28E+08,
        2.96E+08, 2.56E+08, 3.29E+08, 3.36E+08, 3.45E+08, 3.61E+08, 4.15E+08, 4.05E+08,
        4.05E+08, 4.28E+08, 4.02E+08, 4.35E+08, 4.13E+08, 3.78E+08, 4.31E+08, 4.36E+08,
        4.72E+08, 4.70E+08, 5.41E+08, 5.40E+08, 5.04E+08, 5.06E+08, 4.65E+08, 4.95E+08,
        5.14E+08, 4.68E+08, 5.35E+08
    ]
}
df = pd.DataFrame(data)

# Data Preparation
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)
df.sort_index(inplace=True)

# Train-Test Split
train_size = int(len(df) * 0.8)  # 80% for training, 20% for testing
train, test = df[:train_size], df[train_size:]

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(np.arange(len(train)).reshape(-1, 1), train['Emissions'])
lr_predictions = lr_model.predict(np.arange(len(train), len(train) + len(test)).reshape(-1, 1))

# SARIMAX
sarimax_model = SARIMAX(train['Emissions'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarimax_results = sarimax_model.fit()
sarimax_predictions = sarimax_results.get_forecast(steps=len(test)).predicted_mean

# LSTM
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

# Normalize the dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)

look_back = 1
trainX, trainY = create_dataset(train_scaled, look_back)
testX, testY = create_dataset(test_scaled, look_back)

trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

lstm_model = Sequential()
lstm_model.add(LSTM(4, input_shape=(1, look_back)))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mean_squared_error', optimizer='adam')
lstm_model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=2)
lstm_predictions = lstm_model.predict(testX)
lstm_predictions = scaler.inverse_transform(lstm_predictions)

# Evaluation Metrics
lr_mae = mean_absolute_error(test['Emissions'], lr_predictions)
sarimax_mae = mean_absolute_error(test['Emissions'], sarimax_predictions)
lstm_mae = mean_absolute_error(test['Emissions'][look_back:], lstm_predictions)

lr_rmse = sqrt(mean_squared_error(test['Emissions'], lr_predictions))
sarimax_rmse = sqrt(mean_squared_error(test['Emissions'], sarimax_predictions))
lstm_rmse = sqrt(mean_squared_error(test['Emissions'][look_back:], lstm_predictions))

print(f'Linear Regression MAE: {lr_mae}, RMSE: {lr_rmse}')
print(f'SARIMAX MAE: {sarimax_mae}, RMSE: {sarimax_rmse}')
print(f'LSTM MAE: {lstm_mae}, RMSE: {lstm_rmse}')

# Calculate R2 for Linear Regression
lr_r2 = r2_score(test['Emissions'], lr_predictions)
print(f'Linear Regression R2: {lr_r2}')
# Calculate R2 for SARIMAX
sarimax_residuals = test['Emissions'] - sarimax_predictions
sarimax_ss_res = np.sum(sarimax_residuals**2)
sarimax_ss_tot = np.sum((test['Emissions'] - np.mean(test['Emissions']))**2)
sarimax_r2 = 1 - (sarimax_ss_res / sarimax_ss_tot)
print(f'SARIMAX R2: {sarimax_r2}')
# Calculate R2 for LSTM
lstm_residuals = test['Emissions'][look_back:] - lstm_predictions.reshape(-1)
lstm_ss_res = np.sum(lstm_residuals**2)
lstm_ss_tot = np.sum((test['Emissions'][look_back:] - np.mean(test['Emissions'][look_back:]))**2)
lstm_r2 = 1 - (lstm_ss_res / lstm_ss_tot)
print(f'LSTM R2: {lstm_r2}')
# Calculate MSE for Linear Regression
lr_mse = mean_squared_error(test['Emissions'], lr_predictions)
print(f'Linear Regression MSE: {lr_mse}')
# Calculate MSE for SARIMAX
sarimax_mse = np.mean((sarimax_residuals) ** 2)
print(f'SARIMAX MSE: {sarimax_mse}')
# Calculate MSE for LSTM
lstm_mse = np.mean((lstm_residuals) ** 2)
print(f'LSTM MSE: {lstm_mse}')
```
